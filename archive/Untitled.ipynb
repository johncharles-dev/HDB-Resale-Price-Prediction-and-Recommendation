```python
from try2_prediction_helper import (
    get_coordinates,
    load_mappings,
    DistanceCalculator,
    predict_price_try2
)
import joblib

# Load resources
mappings = load_mappings()
distance_calc = DistanceCalculator()
model = joblib.load('Model_Building/xgb_resale_all.joblib')

# Real 2025 transaction
user_input = {
    'block': '112A',
    'street': 'Depot Road',
    'town': 'BUKIT MERAH',
    'flat_type': '4 ROOM',
    'flat_model': 'Model A',
    'floor_area_sqm': 100.0,
    'lease_commence_year': 2006,
    'floor_level': 22,  # Midpoint of 10-12
    'year': 2025,
    'month': 1
}

result = predict_price_try2(user_input, model, mappings, distance_calc)

print("="*60)
print("TEST: Real 2025 Transaction")
print("="*60)
print(f"Input Address: {user_input['block']} {user_input['street']}")
print(f"Town: {user_input['town']}")
print(f"Flat: {user_input['flat_type']}, {user_input['floor_area_sqm']} sqm")
print(f"Floor: {user_input['floor_level']}, Lease: {user_input['lease_commence_year']}")
print("="*60)

if result['success']:
    print(f"\n GEOCODING RESULT:")
    print(f"   Full Address: {result['coordinates']['full_address']}")
    print(f"   Latitude:  {result['coordinates']['latitude']}")
    print(f"   Longitude: {result['coordinates']['longitude']}")
    
    print(f"\n CALCULATED DISTANCES:")
    for k, v in result['distances'].items():
        name = k.replace('distance_to_', '').replace('_km', '').replace('_', ' ').title()
        print(f"   {name}: {v:.4f} km")
    
    print(f"\n PREDICTED PRICE: {result['formatted_price']}")
else:
    print(f" Error: {result['error']}")
```
